```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import f_regression
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import r2_score
import pickle as pk
```

```python
df = pd.read_csv("code.csv")
```

```python
df.head()
```

```python
df.loc[pd.to_numeric(df.Age, errors= "coerce").isna()]
```

```python
df.drop([5000], inplace = True)
```

```python
df
```

```python
df.describe()
```

```python
x = df.drop("Personal Loan", axis = 1)
y = df["Personal Loan"]
```

```python
etc = ExtraTreesClassifier()
```

```python
etc.fit(x, y)
```

```python
frature_imp = etc.feature_importances_
```

```python
series = pd.Series(frature_imp, index = x.columns)
series
```

```python
series.nlargest(8)
```

```python
series.nsmallest()
```

```python
x = x.drop(["Securities Account", "Online", "CreditCard", "ID", "ZIP Code", "Mortgage"], axis = 1)
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.3, random_state= 0)
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
score = r2_score(y_test, y_pred)
```

```python
score
```

```python
with open("personal_lone_prediction", "wb") as file:
    pk.dump(rf,file)
```
